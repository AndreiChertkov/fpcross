```python
import numpy as np
import matplotlib.pyplot as plt
import sys

sys.path.append('./../lib')

from sde import Sde
from solver import Solver
```

```python
# SDE parameters

d = 1
A = np.array([
    [2.0],
])
B = np.array([
    [3.0]
])
x0 = np.array([1.0])

t_min = -3.
t_max = +3.
t_poi = 100

def s_func(x=None, t=None):
    return B@x

def f_func(x=None, t=None):
    return A@x

def fx_func(x=None, t=None):
    return A

def x_func(t, w=2):
    return x0 * np.exp(B*w + (A-0.5*B*B)*t)
```

```python
EQ = Sde()
EQ.set(d, d, s_func, f_func, fx_func, x0, x_func)
EQ.set_desc(t_min, t_max, t_poi)

SL = Solver(EQ)ami
SL.solve_ode()
SL.plot_x()
SL.plot_x_err()
SL.plot_r()
```
