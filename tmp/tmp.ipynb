```python
import sys
import time

import numpy as np
from scipy.linalg import expm as expm
import matplotlib as mpl
import matplotlib.pyplot as plt

import tt

sys.path.append('./../lib')
from intertrain import Intertrain
```

```python
def u_func(x):
    return np.sin(np.pi*x)

def u_func_der1(x):
    return np.pi * np.cos(np.pi*x)

def u_func_der2(x):
    return -1. * np.pi * np.pi * np.sin(np.pi*x)

IT = Intertrain(n=[50], l=[ [-1., 1.] ], with_tt=False)

X = IT.grid()
u = u_func(X).reshape(-1)

u_real = u
u_calc = IT.init(u_func).prep().calc(X).reshape(-1)
        

D1 = IT.dif1()
ud1_real = u_func_der1(X).reshape(-1)
ud1_calc = (D1 @ u).reshape(-1)

D2 = IT.dif2()
ud2_real = u_func_der2(X).reshape(-1)
ud2_calc = (D2 @ u).reshape(-1)

x = X.reshape(-1)

plt.title('u')
plt.plot(x, u_real, c='g', label='Real')
plt.scatter(x, u_calc, c='r', label='Calc')
plt.legend(loc='best')
plt.show()

plt.title('du/dx')
plt.plot(x, ud1_real, c='g', label='Real')
plt.scatter(x, ud1_calc, c='r', label='Calc')
plt.legend(loc='best')
plt.show()

plt.title('d2u/dx2')
plt.plot(x, ud2_real, c='g', label='Real')
plt.scatter(x, ud2_calc, c='r', label='Calc')
plt.legend(loc='best')
plt.show()

e0 = np.linalg.norm(u_real - u_calc) / np.linalg.norm(u_real)
e1 = np.linalg.norm(ud1_real - ud1_calc) / np.linalg.norm(ud1_real)
e2 = np.linalg.norm(ud2_real - ud2_calc) / np.linalg.norm(ud2_real)

print('Number of spatial points : %8d'%IT.n[0])
print('Error u                  : %-12.2e'%e0)
print('Error du/dx              : %-12.2e'%e1)
print('Error d2u/dx2            : %-12.2e'%e2)
```
